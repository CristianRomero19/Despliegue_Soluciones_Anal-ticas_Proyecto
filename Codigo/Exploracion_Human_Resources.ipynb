```python
#Carga de datos
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

data=pd.read_csv("Human_Resources.csv")
print(data.shape)
data.head()
```

La base de datos contiene 35 variables y 1470 registros

```python
#Identificación de valores nulos
data.isnull().sum()
```

```python
#Estaditicas descriptivas variables numéricas
data.describe()
```

```python
#Tipo de variables
data.dtypes
```

```python
#Gráficos para variables categóricas
categoricas=[]
for i in data.columns:
    if data[i].dtype == "object":
        categoricas.append(i)

otras_categoricas = ["Education", "EnvironmentSatisfaction", "JobSatisfaction", "RelationshipSatisfaction"]
categoricas.extend(otras_categoricas)
for column in categoricas:
    plt.figure(figsize=(8, 6))
    data[column].value_counts().plot.pie(autopct='%1.1f%%', startangle=90)
    plt.title(f'Distribution of {column}')
    plt.ylabel('')
    plt.show()
```

```python
#Attrition vs variables categóricas
sns.set(style="whitegrid")

for column in categoricas:
    plt.figure(figsize=(10, 6))
    sns.countplot(data=data, x=column, hue='Attrition', palette='Set1', dodge=True)
    
    plt.title(f'{column} vs Attrition')
    plt.ylabel('Number of Employees')
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()
```

```python
#Matríz de correlación
correlation_matrix=data.corr()
sns.set(style='white')

plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, annot=True, fmt='.2f', cmap='coolwarm', square=True, cbar_kws={"shrink": .8})
plt.title('Correlation Matrix')
plt.show()
```
